```python
import pandas as pd
import numpy as np
```

```python
spine = pd.read_csv('/content/THESPINE.csv')
```

```python
trivial = pd.read_csv('/content/ncaa_trivial_data_cleaned_up.csv')
```

```python
connector = pd.read_excel('/content/connector.xlsx')
```

```python
spine = pd.merge(spine, connector, left_on='h_market', right_on='short_name', how='inner')
```

```python
spine = spine.rename(columns={'long_name':'h_long_name', 'short_name':'h_short_name'})
```

```python
spine = pd.merge(spine, connector, left_on='a_market', right_on='short_name', how='inner')
```

```python
spine = spine.rename(columns={'long_name':'a_long_name', 'short_name':'a_short_name'})
```

```python
data = pd.merge(spine, trivial, left_on='h_long_name', right_on='long_name', how='inner')
```

```python
data = data.drop(columns={'long_name'})
```

```python
data = data.rename(columns={'avg_attendence':'h_avg_attendence', 'm_enroll':'h_m_enroll', 'f_enroll':'h_f_enroll','t_enroll':'h_t_enroll','budget':'h_budget','ppg':'h_ppg','oppg':'h_oppg'})
```

```python
data = pd.merge(data, trivial, left_on='a_long_name', right_on='long_name', how='inner')
```

```python
data = data.drop(columns={'long_name'})
```

```python
data = data.rename(columns={'avg_attendence':'a_avg_attendence', 'm_enroll':'a_m_enroll', 'f_enroll':'a_f_enroll','t_enroll':'a_t_enroll','budget':'a_budget','ppg':'a_ppg','oppg':'a_oppg'})
```

```python
data = data.drop(columns={'h_short_name', 'a_short_name'})
```

```python
data['d_spread'] = data['h_points_game'] - data['a_points_game']
```

```python
data['d_avg_attendence'] = data['h_avg_attendence'] - data['a_avg_attendence']
data['d_m_enroll'] = data['h_m_enroll'] - data['a_m_enroll']
data['d_f_enroll'] = data['h_f_enroll'] - data['a_f_enroll']
data['d_t_enroll'] = data['h_t_enroll'] - data['a_t_enroll']
data['d_budget'] = data['h_budget'] - data['a_budget']
data['d_ppg'] = data['h_ppg'] - data['a_ppg']
data['d_oppg'] = data['h_oppg'] - data['a_oppg']
```

```python
data.to_csv('output_data.csv', index=False)
```
